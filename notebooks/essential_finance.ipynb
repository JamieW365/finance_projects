## Visualizing stock prices over time

```python
# Take close price data for each working day
df_close = df_stocks.loc[:, 'Close'].copy()
df_close.head()
```

```python
df_close.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show;
```

## Normalizing Time Series Data to a Base Value (100)

```python
df_close.head()
```

```python
# Take the first value in the series
df_close.AAPL.iloc[0]
```

```python
# Div the series by the initial value
df_close.AAPL.div(df_close.AAPL.iloc[0]).mul(100)
```

```python
df_close.iloc[0]
```

```python
# We can normalize the entire close dataset
df_norm = df_close.div(df_close.iloc[0]).mul(100)
df_norm.head()
```

```python
# Now all stocks have the same starting point and we can see which ones have grown the most proportionally
df_norm.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show;
```

## The shift() Method

```python
df_close.head()
```

```python
df_aapl = df_close.AAPL.copy().to_frame()
df_aapl
```

```python
# The shift() method can be used to slide data along the time series by a set number of periods
df_aapl.shift(periods = 1)
```

```python
df_aapl['lag1'] = df_aapl.shift(periods = 1)
df_aapl
```

```python
# Now we can see the difference between subsequent close!
df_aapl['diff'] = df_aapl.AAPL.sub(df_aapl.lag1) 
df_aapl
```

```python
# We can convert this into percentage change!
df_aapl["pct_change"] = df_aapl.AAPL.div(df_aapl.lag1).sub(1).mul(100)
df_aapl
```

## The pct_change() method

```python
df_aapl.head()
```

```python
# Instead of calculating manually, we can use the pct_change methond to calculate percentage differences
df_aapl['pct_change2'] = df_aapl.AAPL.pct_change(periods=1).mul(100)
df_aapl
```

```python
df_aapl['pct_change'].equals(df_aapl.pct_change2)
```

```python
# This can be done across periods using resample!
df_aapl.AAPL.resample('ME').last().pct_change(periods=1).mul(100)
```

```python
# What is the variance of daily returns for appl?
daily_var_returns = df_ret.var()
daily_var_returns
```

## Performance Reporting with rolling()

```python
df_sp500 = pd.read_csv("data/SP500.csv", parse_dates= ["Date"], index_col = "Date", usecols= ["Date", "Close"])
```

```python
df_sp500.head()
```

```python
df_sp500.resample("ME", kind = "period").last()
```

```python
# Calculating monthly return percentages

monthly_returns = df_sp500.resample("ME", kind = "period").last().pct_change().dropna()
monthly_returns
```

```python
# Annualized returns over a 3 year rolling period

monthly_returns["Return"] = monthly_returns.rolling(36).mean()*12
monthly_returns
```

```python
# Annualized standard deviation over a 3 year rolling period

monthly_returns["Risk"] = monthly_returns.Close.rolling(36).std()*np.sqrt(12)
monthly_returns
```

```python
# Dropping null values

monthly_returns.dropna(inplace= True)
```

```python
monthly_returns
```

```python
monthly_returns.iloc[:,-2:].plot(figsize = (15,10), fontsize= 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
# Correlation shows that increased risk usually indicates lower return
monthly_returns.iloc[:,-2:].corr()
```

```python
monthly_returns.iloc[:,-2:].plot(kind = "scatter", x = "Risk", y = "Return", figsize = (15,10), fontsize= 15, s = 40)
plt.show();
```

# Performance and Investment Periods / Time Diversification

```python
SP500 = pd.read_csv("data/SP500.csv", parse_dates= ["Date"], index_col = "Date",
                    usecols = ["Date", "Close"])
SP500
```

```python
SP500.info()
```

```python
SP500.plot(figsize = (15,10), fontsize= 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
month_ret = SP500.resample("ME").last().to_period().pct_change().dropna()
month_ret
```

```python
month_ret.columns = ["m_returns"]
```

```python
month_ret.rolling(3 * 12).mean()*12
```

```python
for years in [1, 3, 5, 10, 20]:
    month_ret["{}Y".format(years)] = month_ret.m_returns.rolling(years*12).mean()*12

month_ret.tail()
```

```python
# Holding over a long period of time massively reduces volatility, and hence risk

month_ret.iloc[:,-5:].plot(figsize = (15,40), subplots =True, fontsize= 15, sharey = True)
plt.legend(fontsize = 20)
plt.show()
```

## Simple Returns vs Log Returns

```python
df = pd.DataFrame(index = [2016, 2017, 2018], data = [100, 50, 95], columns = ["Price"])
df
```

```python
simple_returns = df.pct_change().dropna()
simple_returns
```

```python
# We can calculate the average yearly return by taking the average
simple_returns.mean()
```

```python
# We can use this to calculate expected returns
100 * 1.2 * 1.2
```

```python
np.log(df / df.shift(1))
```

```python
# Log returns can be calculated as follows
log_returns = np.log(df / df.shift(1)).dropna()
log_returns
```

```python
log_returns.mean()
```

```python
100 * np.exp(2 * log_returns.mean())
```
