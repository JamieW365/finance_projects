# Essential Finance

## Imports

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```


## Importing Stock Price Data From Yahoo Finance

```python
# Select stocks to pull from yahoo
ticker = ["AAPL", "BA", "KO", "IBM", "DIS", "MSFT" ]
```

```python
df_stocks = yf.download(ticker, start = "2010-01-01", end = "2024-01-01")
```

```python
df_stocks.head()
```

```python
df_stocks.tail()
```

```python
df_stocks.info()
```

```python
df_stocks.to_csv("data/stocks.csv")
```

## Working with multilevel index

```python
df_stocks.columns = df_stocks.columns.to_flat_index()
df_stocks.head()
```

```python
df_stocks.columns = pd.MultiIndex.from_tuples(df_stocks.columns)
df_stocks.head()
```

```python
df_stocks.swaplevel(axis=1).sort_index(axis=1).head()
```

## Visualizing stock prices over time

```python
# Take close price data for each working day
df_close = df_stocks.loc[:, 'Close'].copy()
df_close.head()
```

```python
df_close.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show;
```
