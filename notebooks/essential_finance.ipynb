```python
df_stocks = yf.download(ticker, start = "2010-01-01", end = "2019-02-06")
```

## Working with multilevel index

```python
df_stocks.columns = pd.MultiIndex.from_tuples(df_stocks.columns)
df_stocks.head()
```

```python
df_stocks.swaplevel(axis=1).sort_index(axis=1).head()
```

## Visualizing stock prices over time

```python
# Take close price data for each working day
df_close = df_stocks.loc[:, 'Close'].copy()
df_close.head()
```

```python
df_close.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show;
```

## Normalizing Time Series Data to a Base Value (100)

```python
df_close.head()
```

```python
# Take the first value in the series
df_close.AAPL.iloc[0]
```

```python
# Div the series by the initial value
df_close.AAPL.div(df_close.AAPL.iloc[0]).mul(100)
```

```python
df_close.iloc[0]
```

```python
# We can normalize the entire close dataset
df_norm = df_close.div(df_close.iloc[0]).mul(100)
df_norm.head()
```

```python
# Now all stocks have the same starting point and we can see which ones have grown the most proportionally
df_norm.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show;
```

## The shift() Method

```python
df_close.head()
```

```python
df_aapl = df_close.AAPL.copy().to_frame()
df_aapl
```

```python
# The shift() method can be used to slide data along the time series by a set number of periods
df_aapl.shift(periods = 1)
```

```python
df_aapl['lag1'] = df_aapl.shift(periods = 1)
df_aapl
```

```python
# Now we can see the difference between subsequent close!
df_aapl['diff'] = df_aapl.AAPL.sub(df_aapl.lag1) 
df_aapl
```

```python
# We can convert this into percentage change!
df_aapl["pct_change"] = df_aapl.AAPL.div(df_aapl.lag1).sub(1).mul(100)
df_aapl
```

## The pct_change() method

```python
df_aapl.head()
```

```python
# Instead of calculating manually, we can use the pct_change methond to calculate percentage differences
df_aapl['pct_change2'] = df_aapl.AAPL.pct_change(periods=1).mul(100)
df_aapl
```

```python
df_aapl['pct_change'].equals(df_aapl.pct_change2)
```

```python
# This can be done across periods using resample!
df_aapl.AAPL.resample('ME').last().pct_change(periods=1).mul(100)
```

# Measuring Stock Performance With Mean Return and STD of Returns

```python
df_aapl = df_close.AAPL.copy().to_frame()
df_aapl
```

```python
df_aapl.pct_change().dropna()
```

```python
df_ret = df_aapl.pct_change().dropna()
```

```python
df_ret.info()
```

```python
df_ret.plot(kind='hist', figsize=(12,8), bins=100)
plt.show();
```

```python
# What is the average daily return for apple?
daily_mean_return = df_ret.mean()
daily_mean_return
```

```python
# What is the variance of daily returns for appl?
daily_var_returns = df_ret.var()
daily_var_returns
```

```python
# What is the standard deviation of daily returns?
daily_std_retunrs = df_ret.std()
daily_std_retunrs
```

```python
# What is annual mean return?
annual_mean_returns = df_ret.mean() * 252 # 252 trading days in a year on average
annual_mean_returns
```

```python
# What is annual variance?
annual_var_returns = df_ret.var() * 252
annual_var_returns
```

```python
# What is annual standard deviation?
annual_std_returns = np.sqrt(annual_var_returns)
annual_std_returns
```

# Return and Risk

```python
# Normalized stock returns

df_norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
df_close.head()
```

```python
# Calculating percentage change for each stock
df_ret = df_close.pct_change().dropna()
df_ret
```

```python
# We can pull stats using the describe method
df_ret.describe()
```

```python
# We can take the mean and standard deviation of each stock from here
df_summary = df_ret.describe().T[['mean', 'std']]
df_summary
```

```python
# We can annualize our daily mean and std statistics
df_summary['mean'] *= 252
df_summary['std'] *= np.sqrt(252)
df_summary
```

```python
df_summary.plot(kind='scatter', x='std', y='mean', figsize=(15,12), s=50, fontsize=15)
for i in df_summary.index:
    plt.annotate(i, xy=(df_summary.loc[i, "std"]+0.002, df_summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```
